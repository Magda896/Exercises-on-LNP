```python
from lexical_diversity import lex_div as ld
```

```python
flt = ld.flemmatize(data)
print(flt[:10])
```

```python
#simple TTR
ld.ttr(flt)
```

```python
#Root TTR
ld.root_ttr(flt)
```

```python
#Log TTR
ld.log_ttr(flt)
```

```python
#Mass TTR
ld.maas_ttr(flt)
```

```python
#Mean segmental TTR (MSTTR)
ld.msttr(flt)
```

```python
#Hypergeometric distribution D (HDD)
#A more straightforward and reliable implementation of vocD (Malvern, Richards, Chipere, & Duran, 2004)
#as per McCarthy and Jarvis (2007, 2010).
ld.hdd(flt)
```

```python
#Measure of lexical textual diversity (MTLD)
#Calculates MTLD based on McCarthy and Jarvis (2010).
ld.mtld(flt)
```

```python
#Measure of lexical textual diversity (moving average, wrap)
#Calculates MTLD using a moving window approach. Instead of 
#calculating partial factors, it wraps to the beginning of the text to complete the last factors.
ld.mtld_ma_wrap(flt)
```

```python
#Measure of lexical textual diversity (moving average, bi-directional)
#Calculates the average MTLD score by calculating MTLD in each direction using a moving window approach.
ld.mtld_ma_bid(flt)
```
